```python
import requests
from bs4 import BeautifulSoup

def scrape_links(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    links = soup.find_all('a')  # 'a'タグを持つすべての要素を見つけます

    poem_links = {}
    for link in links:
        if 'href' in link.attrs:  # 'href'属性が存在するかどうかを確認します
            link_text = link.text
            if link_text.isdigit():  # リンクテキストが数字であるかどうかを確認します
                poem_links[int(link_text)] = link['href']  # 詩の番号とリンクを辞書に保存します
            else:
                poem_links[link_text] = link['href']  # リンクテキストが数字でない場合も辞書に保存します

    return poem_links

url = 'https://hyakuninisshu.com/'
poem_links = scrape_links(url)

for key, link in poem_links.items():
    print(f'{key}: {link}')

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

def scrape_links(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    links = soup.find_all('a')  # 'a'タグを持つすべての要素を見つけます

    poem_links = {}
    for link in links:
        if 'href' in link.attrs:  # 'href'属性が存在するかどうかを確認します
            link_text = link.text
            if link_text.isdigit():  # リンクテキストが数字であるかどうかを確認します
                poem_links[int(link_text)] = link['href']  # 詩の番号とリンクを辞書に保存します
            else:
                poem_links[link_text] = link['href']  # リンクテキストが数字でない場合も辞書に保存します

    return poem_links

url = 'https://hyakuninisshu.com/'
poem_links = scrape_links(url)

# 辞書をデータフレームに変換します
df = pd.DataFrame(list(poem_links.items()), columns=['Key', 'Link'])

# "Key"列に"【xxx番】"の形式の文字列が含まれている行を抽出します
df_with_number = df[df['Key'].str.contains("【\d+番】")].copy()  # .copy()を使用して新しいデータフレームを作成します

# "Key"列から番号だけを抽出します
df_with_number.loc[:, 'Key'] = df_with_number['Key'].apply(lambda x: int(re.search(r'\d+', x).group()))  # .locを使用して行と列の位置を明示的に指定します

print(df_with_number)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

def scrape_poem_details(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    details = {}

    # 各項目を抽出します
    details['上の句'] = soup.find(string='上の句').find_next().text
    details['下の句'] = soup.find(string='下の句').find_next().text
    details['歌人'] = soup.find(string='歌人').find_next().text
    details['決まり字'] = soup.find(string='決まり字').find_next().text
    details['決まり字数'] = int(soup.find(string='決まり字数').find_next().text.replace('字', ''))
    details['収載和歌集'] = soup.find(string='収載和歌集').find_next().text
    details['現代語訳'] = soup.find(string=re.compile('【現代語訳】')).find_next().text.strip()
    details['解説'] = soup.find(string=re.compile('【解説】')).find_next().text.strip()

    return details

# 元のデータフレームに詩の詳細を追加します
for index, row in df_with_number.iterrows():
    details = scrape_poem_details(row['Link'])
    for key, value in details.items():
        df_with_number.loc[index, key] = value

print(df_with_number)
```
